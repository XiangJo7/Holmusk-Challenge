## Library Importing

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

## Dataset Importing

```python
bill_amount_df = pd.read_csv('C:/Users/XP393YF/Downloads/bill_amount.csv')
bill_id_df = pd.read_csv('C:/Users/XP393YF/Downloads/bill_id.csv')
clinical_data_df = pd.read_csv('C:/Users/XP393YF/Downloads/clinical_data.csv')
demographics_df = pd.read_csv('C:/Users/XP393YF/Downloads/demographics.csv')
```

## Dataset Joining

```python
bill_df = bill_id_df.merge(bill_amount_df, on = 'bill_id', how = 'left')
clinical_data_df.rename(columns = {'id': 'patient_id'}, inplace = True)
clinical_bill_df = clinical_data_df.merge(bill_df, on = ['patient_id', 'date_of_admission'], how = 'left')
clinical_bill_df = clinical_bill_df.drop_duplicates(subset = ['patient_id'])
final_df = clinical_bill_df.merge(demographics_df, on = 'patient_id', how = 'left')

col_list = final_df.columns
col_list = col_list.sort_values()
final_df = final_df[col_list]

col = final_df.columns
print('Columns with null values in them')
print('')
for i, j in zip(col, final_df.isna().sum()):
    if j != 0:
        print(i,':', j)
print('')

final_df
```

## Data Preprocessing 

```python
final_df = final_df.dropna(subset = ['medical_history_2', 'medical_history_5'])
final_df = final_df.drop(['bill_id', 'patient_id'], axis = 1)
final_df = final_df.reset_index(drop = True)

for i in range(len(final_df)):
    if final_df.loc[i, 'gender'] == 'm':
        final_df.loc[i, 'gender'] = 'Male'
    elif final_df.loc[i, 'gender'] == 'f':
        final_df.loc[i, 'gender'] = 'Female'
for i in range(len(final_df)):
    if final_df.loc[i, 'race'] == 'chinese':
        final_df.loc[i, 'race'] = 'Chinese'
    elif final_df.loc[i, 'race'] == 'India':
        final_df.loc[i, 'race'] = 'Indian'    
for i in range(len(final_df)):        
    if final_df.loc[i, 'medical_history_3'] == 'No':
        final_df.loc[i, 'medical_history_3'] = 0
    elif final_df.loc[i, 'medical_history_3'] == 'Yes':
        final_df.loc[i, 'medical_history_3'] = 1
    elif final_df.loc[i, 'medical_history_3'] == '1':
        final_df.loc[i, 'medical_history_3'] = 1
    elif final_df.loc[i, 'medical_history_3'] == '0':
        final_df.loc[i, 'medical_history_3'] = 0     
for i in range(len(final_df)):
    if final_df.loc[i, 'resident_status'] == 'Singapore citizen':
        final_df.loc[i, 'resident_status'] = 'Singaporean'
        
category_cols = ['gender', 'medical_history_1','medical_history_1', 'medical_history_2', 'medical_history_3', 
                 'medical_history_4', 'medical_history_5', 'medical_history_6', 'medical_history_7', 
                 'preop_medication_1', 'preop_medication_2', 'preop_medication_3', 'preop_medication_4', 
                 'preop_medication_5', 'preop_medication_6', 'race', 'resident_status', 'symptom_1', 
                 'symptom_2', 'symptom_3', 'symptom_4', 'symptom_5']
datetime_cols = ['date_of_admission', 'date_of_birth', 'date_of_discharge']

for i in category_cols:
    final_df[i] = final_df[i].astype('category')
for j in datetime_cols:    
    final_df[j] = final_df[j].astype('datetime64[ns]')
    
final_df['amount'] = final_df['amount'].round(2) 
final_df['age'] = ((datetime.datetime(2021, 12, 1) - final_df['date_of_birth']) / np.timedelta64(1, 'Y')).round(0)
final_df['admission_length'] = (final_df['date_of_discharge'] - final_df['date_of_admission'])
weight_list, height_list = final_df.weight.to_list(), final_df.height.to_list()
bmi_list = []
for i in range(len(final_df)):
    bmi_list.append(round(weight_list[i] / ((height_list[i] * 0.01) ** 2), 1))
final_df['bmi_index'] = bmi_list

col_list = final_df.columns
col_list = col_list.sort_values()
final_df = final_df[col_list]
final_df = final_df.sort_values(by = ['date_of_admission', 'date_of_discharge',
                                      'admission_length', 'amount'])

final_df
```

## Data Analysis

```python
def yearly_graph_maker(df):
    sns.set_style("darkgrid")
    fig, ax = plt.subplots(figsize = (10, 6))
    plot = sns.lineplot(ax = ax, x = df.date_of_admission, y = df.amount, color = 'midnightblue', alpha = 0.6)
    plot.axhline(np.percentile(df.amount, 90), c = 'red', linestyle = '--')
    plt.legend(['Bill Amount', '90th Percentile Point'])
    plt.title('Time Series Plot of Bill Amounts', weight = 'bold').set_fontsize('15')
    plt.xlabel('Date', weight = 'bold').set_fontsize('13')
    plt.ylabel('Bill Amount (SGD)', weight = 'bold').set_fontsize('13')
    plt.show()

def bill_amount_desc_function(df, year):
    print(year)
    print(df.amount.describe().round(2))
    print('')

yearly_graph_maker(final_df)
bill_amount_desc_function(final_df, 'Overall')
```
